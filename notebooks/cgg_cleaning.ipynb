```python
def coord_format_checker(coord, axis):
    regex_dd = {'latitude': r"([-+]?\d{1,2}\.\d+)", 
                'longitude': r"([-+]?\d{1,3}\.\d+)"}

    regex_dms = {'latitude': r"(\d{1,2})° *(\d{1,2})' *(\d{1,2}(\.\d+)?)\" *([NS])", 
                'longitude': r"(\d{1,3})° *(\d{1,2})' *(\d{1,2}(\.\d+)?)\" *([EW])"}

    regex_ddm = {'latitude': r"(\d{1,2})° *(\d{1,2}(\.\d+)?)' *([NS])", 
                'longitude': r"(\d{1,3})° *(\d{1,2}(\.\d+)?)' *([EW])"}

    if re.match(regex_dd[axis], coord):
        return 'DD'
    elif re.match(regex_dms[axis], coord):
        return 'DMS'
    elif re.match(regex_ddm[axis], coord):
        return 'DDM'
    else:
        return 'invalid format'
```
