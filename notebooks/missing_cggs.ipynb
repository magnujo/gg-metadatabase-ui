```python
db = pd.read_csv(r"c:\Users\glj523\Downloads\edna_archive_sample_202410250904.txt", sep="\t", encoding='utf-8', dtype=str)
db["field_sample_id"] = db["field_sample_id"].str.replace("-", "_")
db["field_sample_id"] = db["field_sample_id"].str.replace(" ", "")
db[~db["field_sample_id"].str.match('^CGG_3_\d{6}$')]
db.loc[84, "field_sample_id"] = 'CGG_3_007859'
db = db.rename(columns={"SamplingDate": "ArchiveSamplingDate", "Submitter": "ArchiveSamplingSubmitter", "SubmissionDate": "ArchiveSamplingSubmissionDate"})

db_cgg_ids = db["field_sample_id"].unique()
```

```python
cgg = pd.read_excel(r"n:\SUN-GI-metadb-test\CGG data\CGG_Database_Sediment_Water_240410_PSO.xlsx", sheet_name="Sediment_Water", dtype=str)
cgg = cgg.dropna(subset=["CGG ID"])
cgg["CGG ID"] = cgg["CGG ID"].str.strip()
cgg["CGG ID"] = cgg["CGG ID"].str.replace("CCG", "CGG")
```

```python
cgg = cgg[cgg["CGG ID"].isin(db_cgg_ids)]
merged = cgg.merge(db, how="inner", left_on="CGG ID", right_on="field_sample_id")
merged = merged.drop(columns=["field_sample_id"])
```
