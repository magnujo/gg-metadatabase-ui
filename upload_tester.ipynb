```python
def simple_test(file_path, sheet_name, table):
    table = pd.read_sql(sql='SELECT * from super_simple_db.' + table, con=engine)
    sheet = pd.read_excel(file_path, engine='openpyxl', sheet_name=sheet_name)
    sheet = sheet.dropna(axis='columns', how='all')
    sheet = sheet.dropna(axis='index', how='all')
    
    assert sheet.shape[1] == table.shape[1] -3
    assert sheet.shape[0] == table.shape[0]
```

```python
simple_test(file_path=sheet, sheet_name="Animal_Plant", table='cgg_animal_plant')
```
